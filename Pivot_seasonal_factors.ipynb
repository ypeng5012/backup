```python
#ramp_data=pd.read_excel('Fwy_2017_ramp_data.xlsx')
ramp_data=pd.read_csv('2017.csv')
```

```python
det=ramp_data['det#'].tolist()
det=list(set(det))
factor_month = []
factor_week = []
factor_nohd_month=[]
factor_nohd_week=[]
```

```python
for number in det:
        ramp_data_loc=ramp_data[ramp_data['det#']== number]    
        df=ramp_data_loc[ramp_data_loc.Imputation!='Bad']
        df_nohd=df[df.Holiday_flag!='Holiday']
        ave=pd.pivot_table(df,index='month',columns='week',values="MD",aggfunc='mean', margins=True)
        ave_nohd=pd.pivot_table(df_nohd,index='month',columns='week',values="MD",aggfunc='mean', margins=True)
        count=pd.pivot_table(df,index='month',columns='week',values="MD",aggfunc='count', margins=True)
        count_nohd=pd.pivot_table(df_nohd,index='month',columns='week',values="MD",aggfunc='count', margins=True)
        row_index=['Jan', 'Feb', 'Mar', 'Apr' , 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec','All']
        col_index=['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday','Friday','Saturday','All']
        ave = ave.reindex(col_index, axis = 1)
        ave = ave.reindex(row_index, axis = 0)

        ave_nohd = ave_nohd.reindex(col_index, axis = 1)
        ave_nohd = ave_nohd.reindex(row_index, axis = 0)

        count= count.reindex(col_index, axis = 1)
        count = count.reindex(row_index, axis = 0)

        count_nohd= count_nohd.reindex(col_index, axis = 1)
        count_nohd = count_nohd.reindex(row_index, axis = 0)

        #calculate factors
        factor=ave/ave.loc["All" , "All"]
        factor_nohd=ave_nohd/ave_nohd.loc["All" , "All"]
        
        # Creating Excel Writer Object from Pandas  
       # writer = pd.ExcelWriter(str(number)+'.xlsx',engine='xlsxwriter')   
       # workbook=writer.book
       # worksheet=workbook.add_worksheet('result')
       # writer.sheets['result'] = worksheet
       # factor.to_excel(writer,sheet_name='result',startrow=0 , startcol=0)   
       # count.to_excel(writer,sheet_name='result',startrow=0, startcol=12) 
       # factor_nohd.to_excel(writer,sheet_name='result',startrow=20 , startcol=0)   
      #  count_nohd.to_excel(writer,sheet_name='result',startrow=20, startcol=12) 
       # writer.save()
        
        factor_month.append(factor['All'])
        factor_week_value=factor.iloc[['12']].T
        factor_week.append(factor_week_value['All'])
        
        factor_nohd_month.append(factor_nohd['All'])
        factor_nohd_week_value=factor_nohd.iloc[['12']].T
        factor_nohd_week.append(factor_nohd_week_value['All'])
```

## export

```python
factor_month=pd.DataFrame(factor_month)
factor_month.index=det

factor_week=pd.DataFrame(factor_week)
factor_week.index=det

factor_nohd_month=pd.DataFrame(factor_nohd_month)
factor_nohd_month.index=det

factor_nohd_week=pd.DataFrame(factor_nohd_week)
factor_nohd_week.index=det

# Creating Excel Writer Object from Pandas  
writer = pd.ExcelWriter('MD.xlsx',engine='xlsxwriter')   
workbook=writer.book
worksheet=workbook.add_worksheet('month_factor')
writer.sheets['month_factor'] = worksheet
factor_month.to_excel(writer,sheet_name='month_factor',startrow=0 , startcol=0)   
factor_nohd_month.to_excel(writer,sheet_name='month_factor',startrow=0, startcol=16) 

worksheet=workbook.add_worksheet('week_factor')
writer.sheets['week_factor'] = worksheet
factor_week.to_excel(writer,sheet_name='week_factor',startrow=0 , startcol=0)   
factor_nohd_week.to_excel(writer,sheet_name='week_factor',startrow=0, startcol=11) 

writer.save()
```
